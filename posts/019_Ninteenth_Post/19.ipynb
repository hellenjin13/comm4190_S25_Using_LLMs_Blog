## Introduction

Fitness advice is everywhere, but not all of it is good. I wanted to see if GPT4 could give helpful and safe workout recommendations. Can it build a plan that feels realistic and not overwhelming? Does it know the difference between beginner and advanced routines? And most importantly, does it avoid harmful or unsustainable advice?

To test it, I gave the model a few different fitness goals and looked at how it responded. I was interested in both the structure and the tone of its advice.

## Experiment: Prompting GPT4 for Workout Plans

Here are the three prompts I tested

“Make me a workout plan for someone who is just starting to exercise”  
“I want to build muscle but I only have dumbbells and no gym”  
“Give me a short daily workout I can do in my apartment without equipment”

## Prompt 1: “Make me a workout plan for someone who is just starting to exercise”

**Response:**  
Week 1 to 2  
Walk for 20 to 30 minutes 3 times a week  
Bodyweight squats, wall pushups, and seated leg raises 2 times a week  
Stretching or yoga for 10 minutes on off days

Week 3 to 4  
Add light jogging intervals  
Increase reps of bodyweight exercises  
Add light core work like planks and bird dogs

Start slow and listen to your body. Consistency matters more than intensity at the beginning.

**Analysis:**  
This was a great start. The model broke things into weeks, kept the exercises simple, and focused on building a habit. It also gave helpful reminders about listening to your body. Nothing felt extreme or too fast.

## Prompt 2: “I want to build muscle but I only have dumbbells and no gym”

**Response:**  
Try a 3 day split  
Day 1 upper body  
Dumbbell bench press  
Bent over rows  
Bicep curls  
Overhead press

Day 2 lower body  
Goblet squats  
Lunges  
Romanian deadlifts  
Calf raises

Day 3 full body  
Deadlifts  
Pushups  
Dumbbell snatch or clean  
Core circuit

Rest between days and increase weight or reps over time. Eat enough protein to support muscle growth.

**Analysis:**  
This answer was structured, efficient, and doable with limited equipment. It felt like something a personal trainer might suggest to someone working out at home. It also included a nutrition reminder, which was a nice bonus.

## Prompt 3: “Give me a short daily workout I can do in my apartment without equipment”

**Response:**  
10 bodyweight squats  
10 pushups  
20 second plank  
15 jumping jacks  
Repeat 3 times if you can

This can be done in 15 to 20 minutes and helps build strength and cardio. Add variety by swapping in mountain climbers or glute bridges.

**Analysis:**  
This was short, practical, and space friendly. It was also adaptable for different fitness levels. The tone felt encouraging but not pushy. A good option for people who just want to move a little every day.

## Conclusion

GPT4 can give realistic and safe workout advice if you ask clearly. It understands different fitness levels, works within equipment limits, and avoids extreme language or unsustainable routines. While it is not a personal trainer, it is definitely a useful tool for getting started or staying on track. Just remember to check with a doctor if you have health concerns, and always listen to your own body first.
